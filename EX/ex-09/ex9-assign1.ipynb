```python
import re
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.toyo.ac.jp/toyo2023/basic_data/"
csv_out = 'num_students_campas.csv'
```

```python
s = requests.Session()
r = s.get(url)
r.encoding = r.apparent_encoding
bs = BeautifulSoup(r.text, 'html.parser')
```

```python
try:
    target_table = bs.select('.data-table')[4]
    print("5番目の表を取得しました。")
except IndexError:
    print("エラー: 5番目の表が見つかりませんでした。")
    target_table = None
```

```python
if target_table:
    thead_th = target_table.thead.find_all('th')
    header = [th.text for th in thead_th]
    tbody_obj = target_table.tbody
    tr_list = tbody_obj.find_all('tr')
    
    data_rows = []
    
    for tr in tr_list:
        th = tr.find('th')
        if th:
            name = re.sub(r'（.*）', '', th.text)
            
            tds = tr.find_all('td')
            if tds:
                nums = [td.text for td in tds]
                data_rows.append([name] + nums)

    df = pd.DataFrame(data_rows, columns=header)
    
    print(f"データ件数: {len(df)}")
    display(df)
    
    df.to_csv(csv_out, index=False)
    print(f"ファイル '{csv_out}' を保存しました。")
```
