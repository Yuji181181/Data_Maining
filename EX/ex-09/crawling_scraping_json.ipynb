### Sample program for crawling (JSON)  

#### Import libraries  

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from pandas import json_normalize
```

#### Parameters  

```python
# set your API key
url = 'http://api.openweathermap.org/data/2.5/weather?lat=35.780406&lon=139.715592&appid=9d3699936ffcba4faef1c32427992b25'
```

#### Crawling of web page  

```python
s = requests.Session()
r = s.get(url)
json = r.json()
print(json)
```

#### Dic to dataframe  

```python
df_json = json_normalize(json)
print(df_json.shape)
display(df_json)
print(df_json.columns)
```

#### Get info from df  

```python
weather_list = df_json.at[0, 'weather']
print(weather_list)
weather = weather_list[0]['main']
print(weather) # 天候 weather
```

```python
temp = df_json.at[0, 'main.temp'] # 気温 (単位:Kelvin) temperature (Kelvin) 
print(temp)
```

```python
pres = df_json.at[0, 'main.pressure'] # 気圧 pressure
print(pres)
```

```python
city = df_json.at[0, 'name'] # 場所 city name
print(city)
```
