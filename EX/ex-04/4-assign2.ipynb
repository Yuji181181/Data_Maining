```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("toyo_students.csv", delimiter=",")

print(df.shape)
print(df.info())
display(df.head())
```

```python
df['Sum'] = df[['Male','Female']].sum(axis=1)
display(df)
```

```python
plt.bar(df['Campus'],df['Sum'])
plt.xlabel('Campus')
plt.ylabel('# of students')
```

```python
X = np.array(range(df.shape[0]))

plt.xticks(X,df['Campus'])
plt.bar(X-0.4,df['Male'],width=0.4,label='Male')
plt.bar(X,df['Female'],width=0.4,label='Female')
plt.xlabel('Campus')
plt.ylabel('# of students')
plt.legend()
```

```python
height = np.zeros(4)
X = np.array(range(df.shape[0]))

plt.xticks(X, df['Campus'])
for i,r in df[['Male','Female']].T.iterrows():
    plt.bar(X, r, bottom=height,label=r.name)
    height+=r
plt.xlabel('Campus')
plt.ylabel('# of students')
plt.legend()
```

```python
plt.pie(df['Sum'], autopct="%.1f%%", labels=df['Campus'])
```

```python
explodes = np.zeros(len(df))
explodes[3] = 0.1
plt.pie(df['Sum'], autopct="%.1f%%", shadow=True, 
        startangle=90, counterclock=False,
        explode=explodes,
        labels=df['Campus'])
```
