```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("temp_data.csv")

print(df.shape)
print(df.info())
display(df.head())
```

```python
df['Date']=pd.to_datetime(df['Date'])
display(df.head())
```

```python
plt.plot(df["Date"], df["Tokyo"], label='Tokyo')
plt.plot(df["Date"], df["Naha"], label='Naha')
plt.plot(df["Date"], df["Sapporo"], label='Sapporo')
plt.xlabel("Date")
plt.ylabel("Temperature")
plt.legend()
```

```python
plt.hist(df['Tokyo'], bins=20) 
plt.xlabel("Temperature")
plt.ylabel("Frequency")
```

```python
ser_tokyo = df['Tokyo']
ser_20_ave = ser_tokyo.rolling(window=20).mean()
plt.plot(df["Date"], df["Tokyo"], label='Tokyo') 
plt.plot(df["Date"], ser_20_ave, label='20-ave')
plt.xlabel("Date")
plt.ylabel("Temperature")
plt.legend()
```

```python
plt.scatter(df['Tokyo'],df['Naha'])
plt.xlabel("Tokyo")
plt.ylabel("Naha")
```
