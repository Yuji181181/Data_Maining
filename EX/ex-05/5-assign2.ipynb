```python
import pandas as pd
import folium
```

```python
df = pd.read_csv('population.csv', header=0)
print(df.shape)
print(df.info())
display(df.head())
```

```python
geo = 'japan.json'
m = folium.Map(location=[38,135], zoom_start=5)

folium.Choropleth(
    geo_data=geo,
    key_on='feature.properties.name',
    data=df,
    columns=['Prefecture', 'All'],
    fill_color='BuPu',
    legend_name='Total population'
).add_to(m)

display(m)
```

```python
df['ad'] = df['65_over']/df['All']
m = folium.Map(location=[38,135], zoom_start=5)

folium.Choropleth(
    geo_data=geo,
    key_on='feature.properties.name',
    data=df,
    columns=['Prefecture', 'ad'],
    fill_color='BuPu',
    legend_name='Senior population ratio'
).add_to(m)

display(m)
```
