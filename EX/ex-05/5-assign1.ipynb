```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('user_data.csv', header=0)
print(df.shape)
print(df.info())
print(df.head())
```

```python
print(df[df["gender"]=="M"]["age"].mean())
print(df[df["gender"]=="F"]["age"].mean())
```

```python
print(df["occupation"].value_counts())
```

```python
df2 = df[(df["occupation"]=="artist")&(df["gender"]=="M")]
```

```python
print(df2.loc[df2["age"].idxmin(),"zip_code"])
```

```python
data = df['occupation'].value_counts()
plt.bar(data.index, data, label="Occupation Frequency")
plt.title("Occupation Investigation")
plt.xlabel("Occupations")
plt.ylabel("Frequency")
plt.legend()
plt.xticks(rotation=45, fontsize=9)
```
