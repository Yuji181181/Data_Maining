```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
data='Kuramoto_1.csv'
df = pd.read_csv(data,  skiprows=0, header=0, encoding='shift_jis')
print(df.shape)
print(df.info())
display(df.head())
```

```python
df_wood= df[df['構造']=='木造']
display(df_wood)
```

```python
df2 = df_wood[['建築年代','被害度合','割合']]
display(df2)
```

```python
pivot = pd.pivot_table(df2, values="割合", columns="被害度合", index='建築年代',aggfunc="sum")
display(pivot)
```

```python
height = np.zeros(pivot.shape[0])
X = np.array(range(pivot.shape[0]))
plt.xticks(X, pivot.index)
for i,r in pivot.T.iterrows() :
    plt.bar(X, r, bottom=height, label=r.name)
    height+=r
plt.xlabel('Category')
plt.ylabel('Annual Spending (m.u.)')
plt.xticks(rotation=90)
plt.legend()
```
