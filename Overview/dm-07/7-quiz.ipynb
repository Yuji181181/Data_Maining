```python
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

vocabulary = ["red", "sweetness", "skin", "seeds", "sourness", "orange"]

train_docs = [
    "red sweetness skin seeds",
    "red sweetness sourness sweetness",
    "orange sourness sweetness skin"
]

train_labels = ["Apple", "Strawberry", "Orange"]
test_doc = ["seeds sourness sweetness"]

vectorizer = CountVectorizer(vocabulary=vocabulary)
X_train = vectorizer.fit_transform(train_docs)
X_test = vectorizer.transform(test_doc)
model = MultinomialNB(alpha=1.0, fit_prior=True)
model.fit(X_train, train_labels)

probabilities = model.predict_proba(X_test)
prediction = model.predict(X_test)
probs_for_test_doc = probabilities[0]

for i, class_name in enumerate(model.classes_):
    print(f"{class_name}: {probs_for_test_doc[i] * 100:.1f}%")
```
